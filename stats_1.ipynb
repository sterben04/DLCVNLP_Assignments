```python
# Problem Statement 4 & 9
from scipy.integrate import quad
import math
#p(x>12.6mm)
def integrand(d):
    return 20*math.exp(-20*(d-12.5))
p_grtr_126 =   quad(integrand,12.6,np.inf)  
print(p_grtr_126[0])
proportion = quad(integrand,12.5,12.6)
print(proportion[0])
cdf = quad(integrand,0,11.0)
print(cdf[0])   
# Hence the likelihood of the diameter of part being in between 12.5mm and 12.6mm is very high (0.86)
```
